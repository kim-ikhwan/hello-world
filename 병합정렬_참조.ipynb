```python
arr=list(np.random.randint(0,1000,100))
count=0
def merge_sort(arr):
    global count
    print()
    print('count:',count)
    count +=1
    print('arr:',arr)
    print('len(arr):',len(arr))
    if len(arr)<2:
        print('*'*10)
        return arr
    mid=len(arr)//2
    print('mid:',mid)
    
   
    low_arr=merge_sort(arr[:mid])
    print('low_arr:',low_arr)
    high_arr=merge_sort(arr[mid:])
    
    print('high_arr:',high_arr)
    
    merged_arr=[]
    l=h=0
    print('while_start:','-'*10)
    while l< len(low_arr) and h < len(high_arr):
        if low_arr[l]< high_arr[h]:
            merged_arr.append(low_arr[l])
            l +=1
        else:
            merged_arr.append(high_arr[h])
            h +=1
    print('while_end:','-'*10)
    merged_arr += low_arr[l:]
    merged_arr += high_arr[h:]
    print('merged_arr:',merged_arr)
    print()
    return merged_arr
```
